```python
import re
import requests
from bs4 import BeautifulSoup

url = 'https://www.worldometers.info/coronavirus/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')

table = soup.find(id="main_table_countries_today")
table_body = table.find('tbody')
rows = table_body.find_all('tr')

data = [] 
for row in rows:
    column = [] 
    cols = row.find_all('td')
    for col in cols:
        column.append(col.get_text(strip=True))
    data.append(column)
    
countries = []
total_cases = []
total_deaths = []
total_recovered = []
total_tests = []
population = []
continents = []

for i in range(8, len(data)):
    countries.append(data[i][1])
    total_cases.append(data[i][2])
    total_deaths.append(data[i][4])
    total_recovered.append(data[i][6])                      
    total_tests.append(data[i][11])
    population.append(data[i][13])
    continents.append(data[i][14])
```

```python
import pandas as pd
import numpy as np

columns = {"Country":countries, "Total Cases": total_cases, "Total Deaths": total_deaths, "Total Recovered": total_recovered, "Total Tests": total_tests, "Population": population, "Continent": continents}
data = pd.DataFrame(columns)

data
```
